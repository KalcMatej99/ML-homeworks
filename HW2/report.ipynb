```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("part_2_betas.csv")
df2 = pd.read_csv("part_2_betas_with_uncertainty.csv")
df = df.set_index("ShotType")
df2 = df2.set_index("ShotType")
```

```python
dfc = df[["Competition_U14", "Competition_U16", "Competition_SLO1", "Competition_NBA"]]
labels = df2[["Competition_U14", "Competition_U16", "Competition_SLO1", "Competition_NBA"]].values
print(labels)
plt.figure(figsize = (10,6))

sns.heatmap(dfc, cmap="Spectral",annot = labels,fmt = '')
plt.title("Relationship between competitions and Shot types")
plt.show()
```

```python
dfc = df[["Movement_drive", "Movement_no", "Transition", "TwoLegged","PlayerType_F", "PlayerType_G"]]
labels = df2[["Movement_drive", "Movement_no", "Transition", "TwoLegged","PlayerType_F", "PlayerType_G"]].values

plt.figure(figsize = (11,6))

sns.heatmap(dfc, cmap="Spectral",annot=labels,fmt = '')

plt.title("Relationship between binary features and Shot types")
plt.show()
```

```python
dfc = df[["Angle", "Distance"]]
labels = df2[["Angle", "Distance"]].values

ax = dfc.plot.barh(rot=0)
plt.title("Angle and Distance relationship with Shot types")
plt.show()
```
